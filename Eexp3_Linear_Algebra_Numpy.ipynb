```python
import numpy as np

# Define the coefficient matrix A
A = np.array([[2, 3], [4, 5]])

# Define the constant vector B
B = np.array([10, 20])

# Solve the system of linear equations
x_y = np.linalg.solve(A, B)

# Print the solution
print(f"The solution for x and y is: {x_y}")
print(f"x = {x_y[0]:.2f}")
print(f"y = {x_y[1]:.2f}")
```
