To find the inverse of a matrix, we can use the `numpy.linalg.inv()` function. First, let's create a sample square matrix.

```python
import numpy as np

# Define a sample square matrix
# Note: A matrix must be square and non-singular (determinant not zero) to have an inverse.
matrix_a = np.array([
    [4, 7],
    [2, 6]
])

print("Original Matrix A:")
print(matrix_a)
```

Now, let's calculate its inverse. It's a good practice to check the determinant first to ensure the inverse exists. If the determinant is zero, the matrix is singular and has no inverse.

```python
# Calculate the determinant of the matrix
det_a = np.linalg.det(matrix_a)
print(f"Determinant of Matrix A: {det_a:.4f}")

if det_a == 0:
    print("This matrix is singular and does not have an inverse.")
else:
    # Calculate the inverse of the matrix
    inverse_matrix_a = np.linalg.inv(matrix_a)

    print("\nInverse of Matrix A:")
    print(inverse_matrix_a)

    # Verify the inverse: A * A_inv should be an identity matrix
    identity_check = np.dot(matrix_a, inverse_matrix_a)
    print("\nVerification (A * A_inv):")
    print(identity_check.round(4)) # Round to handle floating point inaccuracies
```
