```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Plot the histogram
fig = plt.figure(figsize=(8, 6))
plt.hist(s, bins=50, density=True, alpha=0.6, color='g')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of a Normal Distribution')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Assuming 's' is the dataset from the previous step
mean_s = np.mean(s)
median_s = np.median(s)
variance_s = np.var(s)
std_dev_s = np.std(s)

print(f"Mean: {mean_s:.4f}")
print(f"Median: {median_s:.4f}")
print(f"Variance: {variance_s:.4f}")
print(f"Standard Deviation: {std_dev_s:.4f}")
```

```python
import numpy as np

# Simulate coin tosses (0 for tails, 1 for heads)
num_tosses = 1000
tosses = np.random.randint(0, 2, num_tosses)

# Count heads and tails
heads = np.sum(tosses == 1)
tails = np.sum(tosses == 0)

# Compute probabilities
prob_heads = heads / num_tosses
prob_tails = tails / num_tosses

print(f"Number of tosses: {num_tosses}")
print(f"Number of heads: {heads}")
print(f"Number of tails: {tails}")
print(f"Probability of heads: {prob_heads:.2f}")
print(f"Probability of tails: {prob_tails:.2f}")
```
